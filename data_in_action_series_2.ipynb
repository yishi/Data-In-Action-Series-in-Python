## Make the model

```python
# load the modules
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import StratifiedKFold

from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, ExtraTreesClassifier
```

```python
# model
logistic = LogisticRegression(random_state=0)
knn = KNeighborsClassifier(weights = 'distance')

svc = SVC(kernel='linear')
lin_svc = LinearSVC()

n_estimators = 100
bag = BaggingClassifier(DecisionTreeClassifier(), n_estimators = n_estimators, random_state=0)
rf = RandomForestClassifier(n_estimators=n_estimators, random_state=0)
ExtraTree = ExtraTreesClassifier(n_estimators=n_estimators, random_state=0)

model_list = [logistic, knn, svc, lin_svc, bag, rf, ExtraTree]
```

**When split the dataset to trainset and testset, random sample is important.**

If the construction of the dataset is uneven, order sample perhaps choose simple construction data as trainset and difficult construction data as testset, then the effect of models will decrease.

If the construction of the dataset is uneven, order sample perhaps choose difficult construction data as trainset and simple construction data as testset, then the effect of models will increase greatly.

So, if the construction of the dataset is uneven, the score of model after order sample will have a  big range.

We could tune the parameter of shuffle in the function of StratifiedKFold in sklearn module in python to choose whether to do random sample.

```python
skf = StratifiedKFold(face.target, n_folds = 3, shuffle=False)
scores_mean = list()
scores_std = list()

for model in model_list:
    this_scores = cross_val_score(model, face.data, face.target, cv = skf)
    scores_mean.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))
    
plt.errorbar(range(7), scores_mean, yerr=scores_std, fmt = 'ob')
plt.xlim(-1, 7)
plt.xticks(range(7), ('logistic', 'knn', 'svc', 'lin_svc', 'bag', 'rf', 'ETree'))
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title(u'The accuracy of recognize human faces (order sample)');
```

```python
cross_val_score(knn, face.data, face.target, cv = skf)
```

```python
skf = StratifiedKFold(face.target, n_folds = 3, shuffle=True)

scores_mean = list()
scores_std = list()

for model in model_list:
    this_scores = cross_val_score(model, face.data, face.target, cv = skf)
    scores_mean.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))
    
plt.errorbar(range(7), scores_mean, yerr=scores_std, fmt = 'ob')
plt.xlim(-1, 7)
plt.xticks(range(7), ('logistic', 'knn', 'svc', 'lin_svc', 'bag', 'rf', 'ETree'))
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title(u'The accuracy of recognize human faces (random sample)');
```

### Tune the parameters

```python
x_train, x_test, y_train, y_test = train_test_split(face.data, face.target, test_size = 0.3, random_state = 0)

skf_t = StratifiedKFold(y_train, n_folds = 3, shuffle=True)
```

```python
logis_para = [{'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]

log_t = GridSearchCV(LogisticRegression(), logis_para, cv=skf_t)
    
log_t.fit(x_train, y_train)

print(log_t.best_params_)

for params, mean_score, scores in log_t.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r" % (mean_score, scores.std() * 2, params))
```

```python
svc_para = [{'kernel': ['linear'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]

svc_t = GridSearchCV(SVC(), svc_para, cv=skf_t)
    
svc_t.fit(x_train, y_train)

print(svc_t.best_params_)

for params, mean_score, scores in svc_t.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r" % (mean_score, scores.std() * 2, params))
```

```python
# model again
logistic = LogisticRegression(C = 100, random_state=0)

lin_svc = SVC(kernel='linear', C = 0.1)

model_list2 = [logistic, lin_svc]
```

```python
scores_mean = list()
scores_std = list()
for model in model_list2:
    this_scores = cross_val_score(model, face.data, face.target, cv = skf)
    scores_mean.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))
    
plt.errorbar(range(2), scores_mean, yerr=scores_std, fmt = 'ob')
plt.xlim(-1, 2)
plt.xticks(range(2), ('logistic', 'lin_svc'))
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title(u'The accuracy of recognize human faces');
```

### Compare the models

```python
x_train, x_test, y_train, y_test = train_test_split(face.data, face.target, test_size = 0.3, random_state = 0)
```

```python
model_list3 = [logistic]
for model in model_list3:
    y_pred = model.fit(x_train, y_train).predict(x_test)
    print('Model:\n\n%s\n' % model)
    print('Classification Report:\n\n%s' % classification_report(y_test, y_pred))
    print("Confusion matrix:\n\n%s\n" % confusion_matrix(y_test, y_pred))
```

**Logistic Regression:**

- There are some human faces hard to recognize (recall < 1.00), which are 4, 7, 25, 34.
