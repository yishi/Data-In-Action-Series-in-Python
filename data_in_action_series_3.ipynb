## Example II: Explore the data

Performs a pixel-wise Vector Quantization (VQ) of an image of the summer palace (China), reducing the number of colors required to show the image from 96,615 unique colors to 64, while preserving the overall appearance quality.

In this example, pixels are represented in a 3D-space and K-means is used to find 64 color clusters. 

In the image processing literature, the codebook obtained from K-means (the cluster centers) is called the color palette. 

For comparison, a quantized image using a random codebook (colors picked up randomly) is also shown.

```python
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from sklearn.metrics import pairwise_distances_argmin
```

```python
# Load the Summer Palace photo
china = load_sample_image("china.jpg")
```

```python
china.shape
```

```python
china[:, :, 2]
```

```python
plt.axis('off')
plt.imshow(china);
```

```python
# Convert to floats instead of the default 8 bits integer coding. Dividing by
# 255 is important so that plt.imshow behaves works well on float data (need to
# be in the range [0-1]
china = np.array(china, dtype=np.float64) / 255
```

##Example II: Make the model

```python
# Load Image and transform to a 2D numpy array.
w, h, d = original_shape = tuple(china.shape)
image_array = np.reshape(china, (w * h, d))
```

```python
image_array.shape
```

```python
n_colors = 64
```

```python
image_array_sample = shuffle(image_array, random_state=0)[:1000]

kmeans = KMeans(n_clusters=n_colors, random_state=0).fit(image_array_sample)

labels = kmeans.predict(image_array)
```

```python
image_array_sample.shape
```

```python
labels.shape
```

```python
kmeans.cluster_centers_.shape
```

```python
def recreate_image(codebook, labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image
```

```python
plt.axis('off')
plt.title('Quantized image (64 colors, K-Means)')
plt.imshow(recreate_image(kmeans.cluster_centers_, labels, w, h));
```

```python
plt.axis('off')
plt.title('Original image (96,615 colors)')
plt.imshow(china);
```

```python
codebook_random = shuffle(image_array, random_state=0)[:n_colors + 1]

labels_random = pairwise_distances_argmin(codebook_random,image_array,axis=0)
```

```python
codebook_random.shape
```

```python
labels_random.shape
```

```python
labels_random
```

```python
plt.axis('off')
plt.title('Quantized image (64 colors, Random)')
plt.imshow(recreate_image(codebook_random, labels_random, w, h));
```
