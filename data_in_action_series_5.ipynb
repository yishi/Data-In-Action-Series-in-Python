# Data In Action Series 5: Regression Models

```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Explore the data 

We use boston house prices dataset as an example data.

Here, the object is to predict the house prices in boston by some ingredients, such as average number of rooms per dwelling, weighted distances to five Boston employment centres, etc.

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
print(boston.DESCR)
```

```python
boston.data.shape
```

```python
boston_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
boston_df['Medv'] = boston.target
boston_df.head()
```

```python
boston_df.plot(subplots = True);
```

From the graph, the construction of data set is unstable.

Using the variable RAD as an example, we could see its value has a big change.

If we choose the stable part, which has the small value relatively, as the train set, and choose the other part as the test set, the performance of predicted model will be worse, because of the different construction of the variables.

```python
boston_df.RAD.plot();
```

```python
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV, LarsCV, LassoLarsCV

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

The parameter of shuffle controls how to choose data points as the train set or test set.

Shuffle is equal to False, means we choose the data by order.

From the below code, the range of performance is from -0.2 to 0.7.

```python
kf1 = KFold(n = boston.data.shape[0], n_folds = 5, shuffle = False)
```

```python
cross_val_score(LinearRegression(), boston.data, boston.target, cv = kf1)
```

Shuffle is equal to True, means we choose the data by random.

From the below code, the range of performance is from 0.5 to 0.7.

So, **when the construction of data set is unstable, we'd better choose the data points by random**.

```python
kf2 = KFold(n = boston.data.shape[0], n_folds = 5, shuffle = True)
```

```python
cross_val_score(LinearRegression(), boston.data, boston.target, cv = kf2)
```

## Make the models 

```python
# model
lr = LinearRegression()
ridge = RidgeCV()
lasso = LassoCV()
net = ElasticNetCV()
lar = LarsCV()
lassolars = LassoLarsCV()

tree = DecisionTreeRegressor(random_state = 0)
knn = KNeighborsRegressor()
svr_l = SVR(kernel = 'linear')
svr_r = SVR()

n_estimators = 100
bag = BaggingRegressor(DecisionTreeRegressor(), n_estimators = n_estimators, random_state = 0)
rf = RandomForestRegressor(n_estimators = n_estimators, random_state = 0)
etree = ExtraTreesRegressor(n_estimators = n_estimators, random_state = 0)
ada = AdaBoostRegressor(DecisionTreeRegressor(), n_estimators = n_estimators, random_state = 0)
gbr = GradientBoostingRegressor(n_estimators = n_estimators, random_state = 0)

model_list = [lr, ridge, lasso, net, lar, lassolars, tree, knn, svr_l, svr_r, bag, rf, etree, ada, gbr]
```

```python
score_mean = list()
score_std = list()
for model in model_list:
    score = cross_val_score(model, boston.data, boston.target, cv = kf2, scoring = 'r2')
    score_mean.append(score.mean())
    score_std.append(score.std())
```

```python
plt.errorbar(range(15), score_mean, yerr=score_std, fmt = 'ob')
plt.xlim(-1, 15)
plt.xticks(range(16), ('lr', 'ridge', 'lasso', 'net', 'lar', 'lalar', 'tree', 'knn', 'svr_l', 'svr_r', 'bag', 'rf', 'ETree', 'ada', 'gb'))
plt.xlabel('Models')
plt.ylabel('R2')
plt.title(u'The R2 of the predicted models');
```

```python
score_mean = list()
score_std = list()
for model in model_list:
    score = cross_val_score(model, boston.data, boston.target, cv = kf2, scoring = 'mean_absolute_error')
    score_mean.append(score.mean())
    score_std.append(score.std())
```

```python
plt.errorbar(range(15), score_mean, yerr=score_std, fmt = 'ob')
plt.xlim(-1, 15)
plt.xticks(range(16), ('lr', 'ridge', 'lasso', 'net', 'lar', 'lalar', 'tree', 'knn', 'svr_l', 'svr_r', 'bag', 'rf', 'ETree', 'ada', 'gb'))
plt.xlabel('Models')
plt.ylabel('Mean Absolute Error')
plt.title(u'The mean abosolute error of the predicted models');
```

```python
score_mean = list()
score_std = list()
for model in model_list:
    score = cross_val_score(model, boston.data, boston.target, cv = kf2, scoring = 'mean_squared_error')
    score_mean.append(score.mean())
    score_std.append(score.std())

plt.errorbar(range(15), score_mean, yerr=score_std, fmt = 'ob')
plt.xlim(-1, 15)
plt.xticks(range(16), ('lr', 'ridge', 'lasso', 'net', 'lar', 'lalar', 'tree', 'knn', 'svr_l', 'svr_r', 'bag', 'rf', 'ETree', 'ada', 'gb'))
plt.xlabel('Models')
plt.ylabel('Mean Squared Error')
plt.title(u'The mean square error of the predicted models');
```
